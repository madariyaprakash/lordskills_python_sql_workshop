***WHILE LOOP***

- Executes as long as condition is True.
- Condition checked before every iteration


```python
i = 1 
while i < 5:
    i += 1
    if i ==3:
        continue
    print(i)
    
```

***FOR LOOP***

Used to iterate over sequences:
- list
- tuple
- string
- range
- dictionary



```python
# iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# iterating using range
for i in range(5):
    print(i)  # 0 1 2 3 4

# iterating over a string
for char in "hello":
    print(char)

# iterating over a dictionary
student = {"name": "Alice", "age": 20}
for key, value in student.items():
    print(f"{key}: {value}")
```

***break vs continue***
- `break` - Terminates the loop immediately. Control moves outside the loop.
- `continue` - Skips the current iteration. Loop continues with next iteration.

```python
# break example - stop when we find 3
for i in range(1, 6):
    if i == 3:
        break
    print(i)  # prints 1, 2

print("---")

# continue example - skip 3
for i in range(1, 6):
    if i == 3:
        continue
    print(i)  # prints 1, 2, 4, 5
```
