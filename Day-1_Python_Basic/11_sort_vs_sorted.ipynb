***SORT VS SORTED***

Both are used to sort elements in Python.

Difference:
- sort()   → Modifies original list (in-place)
- sorted() → Returns a new sorted object (does NOT modify original)


list.sort()

- Works only on lists
- Modifies the original list
- Returns None
- Both supports reverse=True


sorted(iterable)

- Works on any iterable (list, tuple, string, set, etc.)
- Does NOT modify original object
- Returns a new sorted list
