***SORT VS SORTED***

Both are used to sort elements in Python.

Difference:
- sort()   → Modifies original list (in-place)
- sorted() → Returns a new sorted object (does NOT modify original)


list.sort()

- Works only on lists
- Modifies the original list
- Returns None
- Both supports reverse=True

```python
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort()
print(numbers)              # [1, 1, 2, 3, 4, 5, 9] - original modified

numbers.sort(reverse=True)
print(numbers)              # [9, 5, 4, 3, 2, 1, 1]
```


sorted(iterable)

- Works on any iterable (list, tuple, string, set, etc.)
- Does NOT modify original object
- Returns a new sorted list


```python
# list
numbers = [3, 1, 4, 1, 5, 9]
result = sorted(numbers)
print(result)    # [1, 1, 3, 4, 5, 9]
print(numbers)   # [3, 1, 4, 1, 5, 9] - original unchanged
```

```python
# on list of strings
names = ["Charlie", "Alice", "Bob"]
print(sorted(names))               # ['Alice', 'Bob', 'Charlie']
print(sorted(names, reverse=True)) # ['Charlie', 'Bob', 'Alice']
```

```python
# on dictionary - sorts by keys
scores = {"Bob": 85, "Alice": 92, "Charlie": 78}
print(sorted(scores))              # ['Alice', 'Bob', 'Charlie'] - sorted keys

# sort by values using key parameter
print(sorted(scores, key=lambda k: scores[k]))  # ['Charlie', 'Bob', 'Alice']
```

```python
# sorted on tuple and string
t = (3, 1, 2)
print(sorted(t))       # [1, 2, 3] - returns list

s = "python"
print(sorted(s))       # ['h', 'n', 'o', 'p', 't', 'y']
```
