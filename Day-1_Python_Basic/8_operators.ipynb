***OPERATORS IN PYTHON***

Operators are special symbols used to perform operations
on variables and values.

Types of Operators:

1. Arithmetic Operators
2. Comparison Operators
3. Assignment Operators
4. Logical Operators
5. Bitwise Operators
6. Membership Operators
7. Identity Operators


***Arithmetic Operators***
- Used for mathematical calculations.
- +, -, *, /, //, %, **

Important

- / → Always returns float

- // → Removes decimal part

- % → Used to check even/odd

- ** → Exponentiation

```python
a, b = 10, 3
print(a + b)   # 13 - addition
print(a - b)   # 7  - subtraction
print(a * b)   # 30 - multiplication
print(a / b)   # 3.333... - always float
print(a // b)  # 3  - floor division
print(a % b)   # 1  - modulus (remainder)
print(a ** b)  # 1000 - exponentiation
```

***Comparison Operators***
- Used to compare values. Returns Boolean (True/False).
- ==, !=, >, <, >=, <=

```python
a, b = 10, 20
print(a == b)  # False
print(a != b)  # True
print(a > b)   # False
print(a < b)   # True
print(a >= 10) # True
print(a <= 5)  # False
```

***Assignment Operators***
- Used to assign values.
- =, +=, -=, *=, /=, //=, %=, **=

```python
x = 10
x += 5   # x = x + 5
print(x) # 15
x -= 3   # x = x - 3
print(x) # 12
x *= 2   # x = x * 2
print(x) # 24
x //= 4  # x = x // 4
print(x) # 6
x **= 2  # x = x ** 2
print(x) # 36
```

***Logical Operators***
- Used in conditions.
- and, or, not

```python
age = 20
has_id = True

print(age >= 18 and has_id)   # True - both must be True
print(age < 18 or has_id)     # True - at least one must be True
print(not has_id)             # False - reverses the boolean
```

***Membership Operators***
- Used to check membership.
- in, not in

```python
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)      # True
print("mango" not in fruits)  # True

text = "Hello World"
print("World" in text)        # True
```

***Identity Operators***
- Used to check memory location.
- is, is not

```python
a = [1, 2, 3]
b = a          # b points to same object
c = [1, 2, 3]  # c is a different object with same values

print(a is b)      # True  - same memory location
print(a is c)      # False - different objects
print(a == c)      # True  - same values
print(a is not c)  # True
```

***Bitwise Operators***
- Used for binary operations.
- &- AND, |- OR, ^ - XOR, ~ - NOT, << - LEFT SHIFT, >> - RIGHT SHIFT
- Mostly used in low level programming

```python
a = 6
b = 3
print(a & b)   # 2  - AND
print(a | b)   # 7  - OR
print(a ^ b)   # 5  - XOR
print(~a)      # -7 - NOT
print(a << 1)  # 12 - left shift (multiply by 2)
print(a >> 1)  # 3  - right shift (divide by 2)

# The & operator compares each bit and set it to 1 if both are 1, otherwise it is set to 0:

# 6 = 0000000000000110
# 3 = 0000000000000011
# --------------------
# 2 = 0000000000000010
# ====================
```
