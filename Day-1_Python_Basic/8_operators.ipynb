***OPERATORS IN PYTHON***

Operators are special symbols used to perform operations
on variables and values.

Types of Operators:

1. Arithmetic Operators
2. Comparison Operators
3. Assignment Operators
4. Logical Operators
5. Bitwise Operators
6. Membership Operators
7. Identity Operators


***Arithmetic Operators***
- Used for mathematical calculations.
- +, -, *, /, //, %, **

Important

- / → Always returns float

- // → Removes decimal part

- % → Used to check even/odd

- ** → Exponentiation

***Comparison Operators***
- Used to compare values. Returns Boolean (True/False).
- ==, !=, >, <, >=, <=

***Assignment Operators***
- Used to assign values.
- =, +=, -=, *=, /=, //=, %=, **=

***Logical Operators***
- Used in conditions.
- and, or, not

***Membership Operators***
- Used to check membership.
- in, not in

***Identity Operators***
- Used to check memory location.
- is, is not

***Bitwise Operators***
- Used for binary operations.
- &- AND, |- OR, ^ - XOR, ~ - NOT, << - LEFT SHIFT, >> - RIGHT SHIFT
- Mostly used in low level programming

```python
a = 6
b = 3
print(a & b)

# The & operator compares each bit and set it to 1 if both are 1, otherwise it is set to 0:

# 6 = 0000000000000110
# 3 = 0000000000000011
# --------------------
# 2 = 0000000000000010
# ====================

```
