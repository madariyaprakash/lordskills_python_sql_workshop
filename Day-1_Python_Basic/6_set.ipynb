***SET***

A set is an unordered, mutable collection of unique elements.

Characteristics:
- Unordered (no indexing)
- Mutable
- Does NOT allow duplicate values
- Uses curly braces {}

important : If we create set with just curly braces it consider dictionary. So, better to create with an element or use the set() to create empty set.

***Creating a Set***

```python
fruits = {"apple", "banana", "cherry"}
numbers = {1, 2, 3, 4, 5}
empty_set = set()  # use set(), not {} (that creates a dict)
print(fruits)
print(type(empty_set))  # <class 'set'>
```

***Duplicate Removal (Important Concept)***
- When we create a set with duplicate values, it automatically removes the duplicates and keeps only unique values.

```python
numbers = {1, 2, 2, 3, 3, 3, 4}
print(numbers)  # {1, 2, 3, 4} - duplicates removed

# useful to remove duplicates from a list
my_list = [1, 2, 2, 3, 3, 4]
unique = list(set(my_list))
print(unique)  # [1, 2, 3, 4]
```

***Adding Elements***
- .add() Adds single element.

```python
fruits = {"apple", "banana"}
fruits.add("cherry")
print(fruits)  # {'apple', 'banana', 'cherry'}
```

***update()***
- Adds multiple elements.
- Accepts list, tuple, set, etc.

```python
fruits = {"apple", "banana"}
fruits.update(["cherry", "mango", "apple"])  # duplicates ignored
print(fruits)  # {'apple', 'banana', 'cherry', 'mango'}
```

***Removing Elements***
- remove() : Raises error if element is not found (in-place).
- discard() : Does not raise error if element is not found (in-place).
- pop() : Removes and returns an arbitrary element(no assumption as any item can be removed) from the set. Raises error if set is empty.
- clear() : Removes all elements from the set.

`Note: pop() in list can also take positive and negative index whereas in set it doesn't take any argument`


```python
fruits = {"apple", "banana", "cherry"}

fruits.remove("banana")   # removes banana
print(fruits)

fruits.discard("mango")   # no error even though mango doesn't exist
print(fruits)

removed = fruits.pop()    # removes arbitrary element
print(removed)

fruits.clear()
print(fruits)  # set()
```

***Set Operations (Very Important)***
- Sets are mainly used for mathematical operations.

1) ***Union ( | ) or .union()***
- combines all the unique items from both A and B.

```python
A = {1, 2, 3}
B = {3, 4, 5}
print(A | B)          # {1, 2, 3, 4, 5}
print(A.union(B))     # {1, 2, 3, 4, 5}
```

2) ***Intersection ( & ) or .intersaction()***
- It returns the common items between A and B

```python
A = {1, 2, 3}
B = {3, 4, 5}
print(A & B)               # {3}
print(A.intersection(B))   # {3}
```

3) ***Difference ( - ) or .difference()***
- It returns a set that contains elements that are in either set A or set B but not in both.

```python
A = {1, 2, 3}
B = {3, 4, 5}
print(A - B)            # {1, 2} - in A but not in B
print(A.difference(B))  # {1, 2}
```

4) ***Symmetric Difference ( ^ ) or .symmetric_difference()***
- Symmetric Difference is a two-way (commutative) operation

```python
A = {1, 2, 3}
B = {3, 4, 5}
print(A ^ B)                       # {1, 2, 4, 5} - in A or B but not both
print(A.symmetric_difference(B))   # {1, 2, 4, 5}
```

5) ***Subset and Superset - .issubset() and .issuperset()***
- set A is a subset of another set B if all elements of set A are also present in set B. 
- set B is a superset of another set A if set B contains all elements of set A (and potentially more). The superset relationship is the inverse of the subset relationship. 

```python
A = {1, 2}
B = {1, 2, 3, 4}
print(A.issubset(B))    # True - all of A is in B
print(B.issuperset(A))  # True - B contains all of A
```

6) ***Frozen Set (Immutable Set)***

```python
fs = frozenset([1, 2, 3, 4])
print(fs)           # frozenset({1, 2, 3, 4})
# fs.add(5)         # AttributeError - cannot modify frozenset
print(type(fs))     # <class 'frozenset'>
```

***Important Characteristics***
- No indexing (cannot access using s[0]) because it's unordered.
- No duplicates
- Faster membership testing (in operator)
- Elements must be immutable (int, str, tuple allowed) means you can't have list(mutable object) inside set.

```python
s = {1, 2, 3, "hello", (4, 5)}
print(2 in s)    # True - fast membership test
print(10 in s)   # False
# s[0]           # TypeError - no indexing in set
# s.add([1, 2])  # TypeError - list is not hashable (mutable)
```
