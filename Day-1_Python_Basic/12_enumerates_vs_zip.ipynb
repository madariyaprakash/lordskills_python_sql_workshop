***enumerate() and zip() are built-in functions***

used mainly in loops.

- enumerate() → Used to get index + value
- zip()       → Used to combine multiple iterables


***enumerate()***
- Adds a counter (index) to an iterable.
- enumerate(iterable, start=1). If we don't pass start arg, it will take default as 0 if we want to pass different number add the start arg.
- we can use the range() function to iterate the items and with index get the corresponding items but enumerates() is much cleaner way to do so.

***zip()***
- Combines multiple iterables element-wise.
- zip() stops at shortest iterable.
- We can convert the zip() to dictionary
