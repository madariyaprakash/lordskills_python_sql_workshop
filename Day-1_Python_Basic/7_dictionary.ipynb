***DICTIONARY (dict)***

A dictionary is a mutable, unordered collection of key-value pairs.

Characteristics:
- Stores data in key: value format
- Keys must be unique
- Keys must be immutable (int, str, tuple)
- Values can be any data type
- Mutable (can modify after creation)
- To create empty dictionary we can use either {} or dict()

***Important Dictionary Methods***
- keys() - Returns all keys.
- values() - Returns all values.
- items() - Returns key-value pairs as tuples.
- update() - Updates dictionary with another dictionary.
- pop() - Removes key and returns its value.
- popitem() - Removes last inserted key-value pair. Returns the value in tuple
- clear() - Clear all the items of dictionary
- copy() - Creates shallow copy. It means new object is created but nested objects are shared.
