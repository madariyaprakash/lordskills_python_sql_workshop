***DICTIONARY (dict)***

A dictionary is a mutable, unordered collection of key-value pairs.

Characteristics:
- Stores data in key: value format
- Keys must be unique
- Keys must be immutable (int, str, tuple)
- Values can be any data type
- Mutable (can modify after creation)
- To create empty dictionary we can use either {} or dict()


***Creating a Dictionary***

```python
# creating a dictionary
student = {"name": "Alice", "age": 20, "grade": "A"}
empty = {}
empty2 = dict()
print(student)
print(type(student))  # <class 'dict'>
```

***Accessing Values***
- If key does not exist → KeyError
- get() does NOT raise error.

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

print(student["name"])          # Alice
print(student.get("age"))       # 20
print(student.get("city"))      # None - no error
print(student.get("city", "N/A"))  # N/A - default value
# print(student["city"])        # KeyError
```

***Important Dictionary Methods***
- keys() - Returns all keys.
- values() - Returns all values.
- items() - Returns key-value pairs as tuples.
- update() - Updates dictionary with another dictionary.
- pop() - Removes key and returns its value.
- popitem() - Removes last inserted key-value pair. Returns the value in tuple
- clear() - Clear all the items of dictionary
- copy() - Creates shallow copy. It means new object is created but nested objects are shared.


```python
student = {"name": "Alice", "age": 20, "grade": "A"}

print(student.keys())    # dict_keys(['name', 'age', 'grade'])
print(student.values())  # dict_values(['Alice', 20, 'A'])
print(student.items())   # dict_items([('name', 'Alice'), ('age', 20), ('grade', 'A')])

# update
student.update({"city": "NYC", "age": 21})
print(student)

# pop
grade = student.pop("grade")
print(grade)    # A

# popitem - removes last inserted
last = student.popitem()
print(last)     # ('city', 'NYC')

# copy
copy_student = student.copy()
print(copy_student)

# iterating over dictionary
for key, value in student.items():
    print(f"{key}: {value}")
```
