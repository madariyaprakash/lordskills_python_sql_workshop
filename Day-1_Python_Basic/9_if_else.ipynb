***CONDITIONAL STATEMENTS***

Used to make decisions in a program.

Keywords:
- if
- elif
- else


***if-else***
- Executes block only if condition is True
- Indentation is mandatory in Python
- If condition is False → else block runs

***if-elif-else***
- Used when checking multiple conditions
- Only one block executes

***Nested if-else***
- One if inside another if
- Used when condition depends on another condition. eg. if person age is > 18 and citizen True then he/she eligible to vote.

***Single line if condition***
- if statment : return statement

***Ternary operator in Python***
- Syntax: value_if_true if condition else value_if_false
