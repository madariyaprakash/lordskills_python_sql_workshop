***CONDITIONAL STATEMENTS***

Used to make decisions in a program.

Keywords:
- if
- elif
- else


***if-else***
- Executes block only if condition is True
- Indentation is mandatory in Python
- If condition is False → else block runs

```python
age = 18
if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

***if-elif-else***
- Used when checking multiple conditions
- Only one block executes

```python
marks = 75
if marks >= 90:
    print("Grade: A+")
elif marks >= 80:
    print("Grade: A")
elif marks >= 70:
    print("Grade: B")
else:
    print("Grade: C")
```

***Nested if-else***
- One if inside another if
- Used when condition depends on another condition. eg. if person age is > 18 and citizen True then he/she eligible to vote.

```python
age = 20
is_citizen = True

if age >= 18:
    if is_citizen:
        print("Eligible to vote")
    else:
        print("Not a citizen, cannot vote")
else:
    print("Too young to vote")
```

***Single line if condition***
- if statment : return statement

```python
x = 10
if x > 5: print("x is greater than 5")
```

***Ternary operator in Python***
- Syntax: value_if_true if condition else value_if_false

```python
age = 20
status = "adult" if age >= 18 else "minor"
print(status)  # adult

# another example
num = 7
result = "odd" if num % 2 != 0 else "even"
print(result)  # odd
```
