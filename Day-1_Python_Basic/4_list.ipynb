***LIST***

A list is an ordered, mutable collection of elements.

Characteristics:
- Ordered (index-based)
- Mutable (can be modified)
- Allows duplicate values
- Can store mixed data types


***Creating a List***

```python
# creating a list
fruits = ["apple", "banana", "cherry"]
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]
empty = []
print(fruits)
print(numbers)
print(mixed)
```

***Accessing Elements***

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0])   # first element
print(fruits[-1])  # last element
print(fruits[1])   # second element
```

***List Slicing***

```python
numbers = [10, 20, 30, 40, 50]
print(numbers[1:4])   # [20, 30, 40]
print(numbers[:3])    # [10, 20, 30]
print(numbers[2:])    # [30, 40, 50]
print(numbers[::2])   # [10, 30, 50] - every 2nd element
```

***Important List Methods***


1) ***append()***
- Adds element at the end.
- Modifies original list
- Adds only one element

```python
fruits = ["apple", "banana"]
fruits.append("cherry")
print(fruits)  # ['apple', 'banana', 'cherry']
```

2) ***extend()***
- Adds multiple elements.


```python
fruits = ["apple", "banana"]
fruits.extend(["cherry", "mango"])
print(fruits)  # ['apple', 'banana', 'cherry', 'mango']
```

3) ***insert()***
- Insert at specific index.

```python
fruits = ["apple", "banana", "cherry"]
fruits.insert(1, "mango")  # insert at index 1
print(fruits)  # ['apple', 'mango', 'banana', 'cherry']
```

4) ***remove()***
- Removes a specific element.
- Removes first matching value when there are duplicates.

```python
fruits = ["apple", "banana", "cherry", "banana"]
fruits.remove("banana")  # removes first occurrence
print(fruits)  # ['apple', 'cherry', 'banana']
```

5) ***pop()***
- Removes element by index (default last).
- Returns removed value

```python
fruits = ["apple", "banana", "cherry"]
removed = fruits.pop()     # removes last
print(removed)             # cherry
removed2 = fruits.pop(0)   # removes at index 0
print(removed2)            # apple
print(fruits)              # ['banana']
```

6) ***clear()***
- Removes all elements.

```python
fruits = ["apple", "banana", "cherry"]
fruits.clear()
print(fruits)  # []
```

7) ***index()***
- Returns index of first match.

```python
fruits = ["apple", "banana", "cherry"]
print(fruits.index("cherry"))  # 2
```

8. ***count()***
- Counts occurrences.

```python
numbers = [1, 2, 3, 2, 4, 2]
print(numbers.count(2))  # 3
```

9) ***sort()***
- Sorts original list.
- it can take parameter `reverse=True` to arrange in descending order

```python
numbers = [3, 1, 4, 1, 5, 9, 2]
numbers.sort()
print(numbers)              # [1, 1, 2, 3, 4, 5, 9]
numbers.sort(reverse=True)
print(numbers)              # [9, 5, 4, 3, 2, 1, 1]
```

10) ***reverse()***
- Reverses original list.

```python
fruits = ["apple", "banana", "cherry"]
fruits.reverse()
print(fruits)  # ['cherry', 'banana', 'apple']
```

11. ***Built-in functions used with list***
- len(), sum(), max(), min(), sorted()

```python
numbers = [3, 1, 4, 1, 5, 9]
print(len(numbers))     # 6
print(sum(numbers))     # 23
print(max(numbers))     # 9
print(min(numbers))     # 1
print(sorted(numbers))  # [1, 1, 3, 4, 5, 9] - new list, original unchanged
```
