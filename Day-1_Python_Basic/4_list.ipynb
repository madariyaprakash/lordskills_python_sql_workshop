***LIST***

A list is an ordered, mutable collection of elements.

Characteristics:
- Ordered (index-based)
- Mutable (can be modified)
- Allows duplicate values
- Can store mixed data types

***Important List Methods***


1) ***append()***
- Adds element at the end.
- Modifies original list
- Adds only one element

2) ***extend()***
- Adds multiple elements.


3) ***insert()***
- Insert at specific index.

4) ***remove()***
- Removes a specific element.
- Removes first matching value when there are duplicates.

5) ***pop()***
- Removes element by index (default last).
- Returns removed value

6) ***clear()***
- Removes all elements.

7) ***index()***
- Returns index of first match.

8. ***count()***
- Counts occurrences.

9) ***sort()***
- Sorts original list.
- it can take parameter `reverse=True` to arrange in descending order

10) ***reverse()***
- Reverses original list.

11. ***Built-in functions used with list***
- len(), sum(), max(), min(), sorted()
