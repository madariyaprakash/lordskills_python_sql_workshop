***INPUT AND OUTPUT***

Input  → Taking data from user
Output → Displaying data to user

Python uses:
- input()  → for taking user input
- print()  → for displaying output


***Input in Python***

***Formatted Output (Very Important)***
- f-string (Recommended)
- format() Method
- Percentage Formatting (Old Method)

```python
a = 2
b = 3 
print(f"there are two numbers {a} and {b} and their sum is {a+b}")
print("there are two numbers {} and {} and their sum is {}".format(a,b,a+b))
print("there are tow numbers %d and %d and their sum is %d" %(a,b,a+b))
```
