
***PYTHON VARIABLES***

A variable is a name that stores a value in memory.

In Python:
- No need to declare data type explicitly.
- Type is decided automatically at runtime.
- Python is dynamically typed.

Syntax:
variable_name = value


```python
# define variable
a = 10
b = 20
# print variable
print("The value of a is:", a)
```


***VARIABLE NAMING RULES***

1. Must start with a letter (a-z, A-Z) or underscore (_)
2. Cannot start with a number
3. Can contain letters, numbers, underscore
4. Case-sensitive (age and Age are different)
5. Cannot use Python keywords (if, for, while, class, etc.)


```python

# valid variable names
age = 30
_name = "Alice"
# invalid variable names
# 1st_name = "Bob"  # starts with a number
# for = 10  # 'for' is a keyword
# print valid variables
print("Age:", age)
```

***MULTIPLE VARIABLE ASSIGNMENT***

```python
# multiple variable assignment
x, y, z = 1, 2, 3
print("x:", x, "y:", y, "z:", z)
```

***CONSTANTS IN PYTHON***
- Python does not have true constants.
- By convention, variables written in CAPITAL letters
are treated as constants.

```python
# constants in Python
PI = 3.14159
GRAVITY = 9.81
print("PI:", PI, "Gravity:", GRAVITY)
```

***GLOBAL VARIABLE***
- Defined outside any function.
- Accessible anywhere in the code.
- Can be modified inside functions using 'global' keyword.

```python
# Global variable
global_var = "I am a global variable"
def my_function():
    # Local variable
    local_var = "I am a local variable"
    print(local_var)
    print(global_var)  # Accessing global variable inside function

my_function()
```
