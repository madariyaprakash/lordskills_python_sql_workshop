***LAMBDA FUNCTION***

A lambda function is a small anonymous function.

- It has no name.
- It can have multiple arguments.
- It can have only ONE expression.
- It automatically returns the result of that expression.

Syntax:
`lambda arguments : expression`


```python
# basic function for addition
def add(a, b):
    return a + b

print(add(3, 5))  # 8
```

```python
# same above function using lambda
add = lambda a, b: a + b
print(add(3, 5))  # 8
```

```python
# lambda with no argument
greet = lambda: "Hello World"
print(greet())  # Hello World
```

```python
# lambda inside another function
"""
This is called a closure. A closure is a function object that has access to variables in its enclosing scope, even after the outer function has finished executing. 
In this example, the inner lambda function has access to the variable 'x' defined in the outer function 'outer_func'. When we call 'add(5)', it adds 5 to the value of 'x' (which is 10) and returns 15.
"""
def outer_func(x):
    return lambda y : x + y
add = outer_func(10)
print(type(add)) # <class 'function'>
print(add(5)) # 15
```
