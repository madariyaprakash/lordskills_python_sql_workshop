***COPYING A DICTIONARY***

There are two ways to copy a dictionary:

1. Shallow Copy  → copies only outer object
2. Deep Copy     → copies outer + all inner objects

`Note: Same applies to other iterable objects such as list, tuple and set`

***Shallow Copy***
- creates new object
- nested object are shared

```python
import copy

original = {"name": "Alice", "scores": [85, 90, 95]}
shallow = copy.copy(original)

# modifying nested list affects both
shallow["scores"].append(100)
print(original["scores"])  # [85, 90, 95, 100] - also changed!
print(shallow["scores"])   # [85, 90, 95, 100]

# modifying top-level key does NOT affect original
shallow["name"] = "Bob"
print(original["name"])    # Alice - unchanged
```

***Deep Copy***
- Deep copy copies everything recursively.
- Everything is copied separately.
- Original dictionary remains unchanged.

```python
import copy

original = {"name": "Alice", "scores": [85, 90, 95]}
deep = copy.deepcopy(original)

# modifying nested list does NOT affect original
deep["scores"].append(100)
print(original["scores"])  # [85, 90, 95] - unchanged
print(deep["scores"])      # [85, 90, 95, 100]
```
