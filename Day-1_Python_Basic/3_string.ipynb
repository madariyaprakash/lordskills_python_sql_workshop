***STRING (str)***

A string is a sequence of characters enclosed in:
- Single quotes  -> 'Hello'
- Double quotes  -> "Hello"
- Triple quotes  -> '''Hello''' or """Hello"""

Strings are:
- Ordered
- Immutable
- Iterable


***Creating Strings***

```python
# create string
my_string = "Hello, World!"
# print string
print(my_string)
```

***Accessing Characters (Indexing)***
- Index starts from 0
- Negative index starts from -1 (last character)

```python
name = "prakash"
for i in range(-1, -len(name)-1, -1):
    print(name[i])
```

***String Slicing***
- Syntax: string[start:stop:step]

```python
#string slicing
my_string = "Hello, World!"
# slice string from index 0 to 4
sliced_string = my_string[0:5]
print(sliced_string)
# slice string from index 7 to the end
sliced_string = my_string[7:]
print(sliced_string)
# slice string from the beginning to index 4
sliced_string = my_string[:5]
print(sliced_string)
# slice string with step
sliced_string = my_string[::2]
print(sliced_string)
# string concatenation
string1 = "Hello"
string2 = "World"
concatenated_string = string1 + ", " + string2 + "!"
print(concatenated_string)
```

***Immutability (Very Important)***

`text = "Hello"`

`text[0] = "h"   ❌ Error (cannot modify string)`

```python
# immutability example
name = "prakash"
#ame[0] = "P"  # This will raise a TypeError because strings are immutable
# To change the string, you need to create a new string
name = "P" + name[1:]  # This creates a new string with the desired change
print(name)  # Output: "Prakash"
# string concatenation example
first_name = "Prakash"
last_name = "Shrestha"
full_name = first_name + " " + last_name  # Concatenating strings with a space in between
print(full_name)  # Output: "Prakash Shrestha"
# string formatting example
age = 30
greeting = f"My name is {full_name} and I am {age} years old."  # Using f-string for formatting
print(greeting)  # Output: "My name is Prakash Shrestha and I am 30 years old."
```

***Important Built-in String Methods***
- upper() and lower()
- strip()
- replace()
- split()
- join()
- find()
- len() (not a method, but a built-in function)
- count()
- startswith() and endswith()
- isalpha() - all alphabet, isdigit() - all digit, isalnum() - all alphanumeric
- sort()



```python
# Example of all the built-in string methods

# 1. capitalize()
text = "hello world"
print(text.capitalize())  # Output: "Hello world"
# 2. casefold()
text = "HELLO WORLD"
print(text.casefold())  # Output: "hello world"
# 3. center()
text = "hello"
print(text.center(20))  # Output: "       hello        "
# 4. count()
text = "hello world"
print(text.count("o"))  # Output: 2
# 5. encode()
text = "hello world"
encoded_text = text.encode("utf-8")
print(encoded_text)  # Output: b'hello world'
# 6. endswith()
text = "hello world"
print(text.endswith("world"))  # Output: True
# 7. expandtabs()
text = "hello\tworld"
print(text.expandtabs(4))  # Output: "hello   world"
# 8. find()
text = "hello world"
print(text.find("o"))  # Output: 4
# 9. format()
text = "Hello, {}!"
print(text.format("Alice"))  # Output: "Hello, Alice!"
# 10. format_map()
text = "Hello, {name}!"
print(text.format_map({"name": "Bob"}))  # Output: "Hello, Bob!"
# 11. index()
text = "hello world"
print(text.index("o"))  # Output: 4
# 12. isalnum()
text = "hello123"
print(text.isalnum())  # Output: True
# 13. isalpha()
text = "hello"
print(text.isalpha())  # Output: True
# 14. isdecimal()
text = "12345"
print(text.isdecimal())  # Output: True
# 15. isdigit()
text = "12345"
print(text.isdigit())  # Output: True
# 16. isidentifier()
text = "variable_name"
print(text.isidentifier())  # Output: True
# 17. islower()
text = "hello"
print(text.islower())  # Output: True
# 18. isnumeric()
text = "12345"
print(text.isnumeric())  # Output: True
# 19. isprintable()
text = "hello world"
print(text.isprintable())  # Output: True
# 20. isspace()
text = "   "
print(text.isspace())  # Output: True
# 21. istitle()
text = "Hello World"
print(text.istitle())  # Output: True
# 22. isupper()
text = "HELLO"
print(text.isupper())  # Output: True
```
