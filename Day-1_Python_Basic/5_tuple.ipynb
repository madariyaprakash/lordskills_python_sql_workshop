
***TUPLE***

A tuple is an ordered, immutable collection of elements.

Characteristics:
- Ordered (index-based)
- Immutable (cannot modify after creation)
- Allows duplicate values
- Can store mixed data types


***Creating a Tuple***

```python
# creating tuples
fruits = ("apple", "banana", "cherry")
numbers = (1, 2, 3, 4, 5)
single = (5,)  # single element tuple needs comma
empty = ()
print(fruits)
print(single)
print(type(single))  # <class 'tuple'>

# impt : a = (5) will return an int.
a = (5)
print(type(a))  # <class 'int'>

# imp : To add element in existing tuple we can use + operator to concatenate two tuples. But it will create a new tuple and assign it to the variable. It will not change the existing tuple. Tuples are immutable.
t1 = (1, 2, 3)
t1 = t1 + (4,)
print(t1)  # (1, 2, 3, 4)
```

***Accessing Elements***

```python
fruits = ("apple", "banana", "cherry")
print(fruits[0])   # apple
print(fruits[-1])  # cherry
print(fruits[1])   # banana
```

***Tuple Slicing***

```python
numbers = (10, 20, 30, 40, 50)
print(numbers[1:4])   # (20, 30, 40)
print(numbers[:3])    # (10, 20, 30)
print(numbers[2:])    # (30, 40, 50)
```

***Immutability (Very Important)***
- Tuples cannot be modified after creation.
- t1 = (1,2,3,4)
- t1[0] = 10 # This will raise an error

```python
t1 = (1, 2, 3, 4)
# t1[0] = 10  # TypeError: 'tuple' object does not support item assignment
print("Tuples are immutable - cannot modify elements")
```

***Tuple Methods (Very Few Because Immutable)***
- count() : Returns the number of occurrences of a specified value.
- index() : Returns the index of the first occurrence of a specified value.
- Other built-in functions also works with tuple - len(), sum(), min(), max(), sorted()

```python
numbers = (1, 2, 3, 2, 4, 2)
print(numbers.count(2))   # 3
print(numbers.index(3))   # 2
print(len(numbers))       # 6
print(sum(numbers))       # 14
print(max(numbers))       # 4
print(min(numbers))       # 1
```

***Tuple Packing and Unpacking (Important Concept)***

```python
# packing
person = "John", 25, "Engineer"  # tuple packing
print(person)  # ('John', 25, 'Engineer')

# unpacking
name, age, job = person
print(name)  # John
print(age)   # 25
print(job)   # Engineer

# extending unpacking
numbers = (1, 2, 3, 4, 5)
first, *middle, last = numbers
print(first)   # 1
print(middle)  # [2, 3, 4]
print(last)    # 5
```

***Why Use Tuple?***
- Tuples are faster than lists due to immutability.
- They can be used as keys in dictionaries (since they are immutable).
- They provide a way to ensure data integrity (since they cannot be modified).
- Used to return multiple values from function

```python
# tuple as dictionary key
locations = {(10, 20): "Point A", (30, 40): "Point B"}
print(locations[(10, 20)])  # Point A

# returning multiple values from function
def get_user():
    return "Alice", 30, "Engineer"

name, age, job = get_user()
print(name, age, job)  # Alice 30 Engineer
```

***Tuple Inside Mutable Object***
- Tuple is immutable
- But if it contains mutable object (like list), that inner object can change

```python
t = (1, 2, [3, 4])
print(t)  # (1, 2, [3, 4])

# cannot modify tuple itself
# t[0] = 10  # Error

# but can modify the list inside tuple
t[2].append(5)
print(t)  # (1, 2, [3, 4, 5])
```
