***MATCH STATEMENT (Structural Pattern Matching)***

Introduced in Python 3.10.

match is similar to switch-case in other languages,
but much more powerful.
- official link - https://docs.python.org/3/tutorial/controlflow.html#match-statements


```python
# Syntax:

# match variable:
#     case pattern1:
#         # code
#     case pattern2:
#         # code
#     case _:
#         # default case
```

***basic example of match statement***


***Combine Values***
- we can pass case like case 1 | 2 | 3 | 4 | 5:

***Matching type***
- We can check the type of data in each case like `case int()` or `case str()`

***Match with Conditions (Guard)***
- if inside case is called a guard. e.g. (case x if x >=18:)

***Matching Lists (Powerful Feature)***
- Can extract values from structure
- We can check if list [1,2] matches in each case.
