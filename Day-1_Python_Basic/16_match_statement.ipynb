***MATCH STATEMENT (Structural Pattern Matching)***

Introduced in Python 3.10.

match is similar to switch-case in other languages,
but much more powerful.
- official link - https://docs.python.org/3/tutorial/controlflow.html#match-statements


```python
# Syntax:

# match variable:
#     case pattern1:
#         # code
#     case pattern2:
#         # code
#     case _:
#         # default case
```

***basic example of match statement***


```python
day = "Monday"

match day:
    case "Monday":
        print("Start of the work week")
    case "Friday":
        print("End of the work week")
    case "Saturday" | "Sunday":
        print("Weekend!")
    case _:
        print("Midweek day")
```

***Combine Values***
- we can pass case like case 1 | 2 | 3 | 4 | 5:

```python
status_code = 404

match status_code:
    case 200 | 201:
        print("Success")
    case 400 | 404:
        print("Client Error")
    case 500 | 503:
        print("Server Error")
    case _:
        print("Unknown status")
```

***Matching type***
- We can check the type of data in each case like `case int()` or `case str()`

```python
value = 42

match value:
    case int():
        print("It's an integer")
    case str():
        print("It's a string")
    case float():
        print("It's a float")
    case _:
        print("Unknown type")
```

***Match with Conditions (Guard)***
- if inside case is called a guard. e.g. (case x if x >=18:)

```python
age = 25

match age:
    case x if x < 13:
        print("Child")
    case x if x < 18:
        print("Teenager")
    case x if x >= 18:
        print("Adult")
```

***Matching Lists (Powerful Feature)***
- Can extract values from structure
- We can check if list [1,2] matches in each case.

```python
data = [1, 3]

match data:
    case [1, 2]:
        print("List contains 1 and 2")
    case [1, x]:
        print("Starts with 1, second value is", x)
    case _:
        print("No match")
```

***Matching Tuple***

```python
point = (0, 5)

match point:
    case (0, y):
        print("Y-axis at", y)
    case (x, 0):
        print("X-axis at", x)
    case _:
        print("Somewhere else")
```
