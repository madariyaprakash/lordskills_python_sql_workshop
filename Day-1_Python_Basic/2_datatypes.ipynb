
***PYTHON DATA TYPES***

Data type defines the type of value a variable can store.

Python has the following built-in data types:

1. Numeric Types:
   - int
   - float
   - complex (real + imaginary part[unit  j])

2. Text Type:
   - str

3. Sequence Types:
   - list
   - tuple
   - range

4. Set Types:
   - set
   - frozenset

5. Mapping Type:
   - dict

6. Boolean Type:
   - bool

7. None Type:
   - NoneType


```python
# define all built-in data types variables
# int
a = 10
# float
b = 3.14
# str
c = "Hello, World!"
# bool
d = True
# list
e = [1, 2, 3, 4, 5]
# tuple
f = (1, 2, 3, 4, 5)
# set
g = {1, 2, 3, 4, 5}
# dict
h = {"name": "Alice", "age": 30, "city": "New York"}
# NoneType
i = None
# complex
j = 1 + 2j
# bytes
k = b"Hello, World!"
# bytearray
l = bytearray(b"Hello, World!")
# memoryview
m = memoryview(b"Hello, World!")
# range
n = range(10)
# frozenset
o = frozenset([1, 2, 3, 4, 5])
# bool
p = False


# print all variables
print("int:", a)
print("float:", b)
print("str:", c)
print("bool:", d)
print("list:", e)
print("tuple:", f)
print("set:", g)
print("dict:", h)
print("NoneType:", i)
print("complex:", j)
print("bytes:", k)
print("bytearray:", l)
print("memoryview:", m)
print("range:", n)
print("frozenset:", o)
print("bool:", p)

```
