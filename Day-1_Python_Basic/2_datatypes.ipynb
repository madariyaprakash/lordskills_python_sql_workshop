
***PYTHON DATA TYPES***

Data type defines the type of value a variable can store.

Python has the following built-in data types:

1. Numeric Types:
   - int
   - float
   - complex (real + imaginary part[unit  j])

2. Text Type:
   - str

3. Sequence Types:
   - list
   - tuple
   - range

4. Set Types:
   - set
   - frozenset

5. Mapping Type:
   - dict

6. Boolean Type:
   - bool

7. None Type:
   - NoneType
