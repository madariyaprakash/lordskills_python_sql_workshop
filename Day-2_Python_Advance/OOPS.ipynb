# inheritance
- Inheritance allows a class (child) to reuse and extend the functionality of another class (parent).
- The child class can override methods of the parent class to provide specific implementations.
- 


````python
class A:
    def test(self):
        print("this is class A")

class B(A):
    def test(self):
        print("this is class b")

class C(A):
    def test(self):
        print("this is class c")

b = B()
b.test()
````

- real world example of inheritance
```
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. Balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"{amount} withdrawn. Balance: {self.balance}")
        else:
            print("Insufficient funds")


# subclass
class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0, interest_rate=0.05):
        super().__init__(owner, balance)
        self.interest_rate = interest_rate

    # ✅ extend functionality
    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest added: {interest}. Balance: {self.balance}")


class CheckingAccount(BankAccount):
    def __init__(self, owner, balance=0, overdraft_limit=500):
        super().__init__(owner, balance)
        self.overdraft_limit = overdraft_limit

    # ✅ override withdraw method
    def withdraw(self, amount):
        if amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"{amount} withdrawn. Balance: {self.balance}")
        else:
            print("Overdraft limit exceeded")

s1 = SavingsAccount("prakash", balance=25000)
s1.withdraw(50)
s1.add_interest()
```

- When you inherit a parent class into a child class, the child automatically gets access to all the parent’s attributes and methods. You don’t need to override them unless you want to change their behavior.


# Abstraction
- Abstraction hides implementation details and exposes only essential features. In Python, this is often done using the abc module.


````python
from abc import ABC, abstractmethod

class Vehicle(ABC): # abstract base class
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("car engine starts with key")

class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine starts with kick")
    
c = Car()
c.start_engine()
b = Bike()
b.start_engine()
````

````python
# real world example
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

    @abstractmethod
    def refund(self, amount):
        pass

class PayPal(PaymentGateway):
    def pay(self, amount):
        print(f"Paid {amount} via PayPal")

    def refund(self, amount):
        print(f"Refunded {amount} via PayPal")

class Stripe(PaymentGateway):
    def first_method(num):
        print(f"This is first method{num}")

    def pay(self, amount):
        print(f"Paid {amount} via Stripe")

    def refund(self, amount):
        print(f"Refunded {amount} via Stripe")


p1 = PayPal()
p1.pay(20000)
p1.refund(20000)

s1 = Stripe()
s1.first_method()
s1.pay(30000)
s1.refund(30000)
````


- The pass here is not the “implementation.”
- It’s simply a placeholder — the abstract method declares that this method must exist, but leaves the actual logic to subclasses.
- In other words, the abstract class is saying: “Any class that inherits from me must provide its own version of pay().”
- A subclass must implement all abstract methods first (otherwise it can’t be instantiated). After that we can add extra methods.


# Polymorphism
- Polymorphism = “many forms”.
- It allows different classes to define the same method name but with different implementations.
- The benefit: you can write code that works on objects of different types without caring about their specific class


- real world example of polymorphism
```

from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

    @abstractmethod
    def refund(self, amount):
        pass

class PayPal(PaymentGateway):
    def pay(self, amount):
        print(f"Paid {amount} via PayPal")

    def refund(self, amount):
        print(f"Refunded {amount} via PayPal")

class Stripe(PaymentGateway):
    def pay(self, amount):
        print(f"Paid {amount} via Stripe")

    def refund(self, amount):
        print(f"Refunded {amount} via Stripe")

class RazorPay(PaymentGateway):
    def pay(self, amount):
        print(f"Paid {amount} via RazorPay")

    def refund(self, amount):
        print(f"Refunded {amount} via RazorPay")

# Polymorphism in action
def checkout(gateway:PaymentGateway, amount):
    gateway.pay(amount)

for g in [PayPal(), Stripe(), RazorPay()]:
    checkout(g,100)

```

# Encapsulation
- Encapsulation = “wrapping data and methods together, and restricting direct access to internal details.”
- It protects the internal state of an object and only exposes controlled ways to interact with it.
- To make the attribute or method private, we use `__` as prefix.

````python
class Bank:
    def __init__(self, owner, balance = 0):
        self.owner = owner
        self.__balance = balance
    
    def print_info(self):
        print(f"owner is {self.owner} and balance is {self.__balance}")
    
    def __private_method(self):
        print("this is private method")


b = Bank("prakash", 1500)
b.owner
b.__private_method()
````

- real world example
```
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance   # private attribute

    def deposit(self, amount):
        self.__balance += amount
        self.__log_transaction("Deposit", amount)  # ✅ private method used internally

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            self.__log_transaction("Withdraw", amount)  # ✅ private method used internally
        else:
            print("Insufficient funds")

    def get_balance(self):
        return self.__balance

    # 🔒 private method
    def __log_transaction(self, action, amount):
        print(f"[LOG] {action} of {amount} by {self.owner}")
```
