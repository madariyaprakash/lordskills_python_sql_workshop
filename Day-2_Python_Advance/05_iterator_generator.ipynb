***Iterator***
- In Python, an iterator is an object that lets you traverse through elements of a collection one at a time. 
- `__iter__()` - returns the iterator object itself
- `__next__()` - returns the next element

```python
mystr = "prakash"
myiter = iter(mystr)
print(next(myiter))
print(next(myiter))
print(next(myiter))
```
