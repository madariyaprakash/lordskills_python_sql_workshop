***Dictionary Comprehension***
- It should be generating dictionary from iterable object (list, tuple, set)

```python
dict_result = {i:i**2 for i in range(1,10) if i%2==0}
dict_result
```

```python
students ={"john":85, "bob":95, "mathew":65}
passed_students = {key:value for key,value in students.items() if value>80}
passed_students
```
