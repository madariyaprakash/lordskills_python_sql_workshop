***Tuple comprehension***
- Python does not provide direct tuple comprehension.
- If we use just parenthesis, it will return the generator object.
- To generate the tuple, we have to use tuple().

```python
# create tuple comprehension
squares = (x**2 for x in range(10))
print(squares)  # This will print a generator object
```

Note: When we create tuple comprehension, we have to use tuple().

```python
squares = tuple(x**2 for x in range(10))
print(squares)  # This will print a generator object
```
