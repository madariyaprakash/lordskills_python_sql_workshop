# Decorators
-  decorators are a powerful feature that let you modify or extend the behavior of functions or methods without changing their actual code.
- A decorator is essentially a function that takes another function as input and returns a new function with extra functionality.
- You apply a decorator using the @decorator_name syntax above a function definition
- Common uses: logging, authentication, measuring execution time, caching, etc.


```python
def decorator(func):
    def wrapper():
        print("before called")
        func()
        print("aftrer called")
    return wrapper

@decorator
def main():
    print("Hello world")

main()
```

```python
# check voter based on age

def check_voter(func):
    def wrapper(age):
        if age >= 18:
            func(age)
        else:
            print("You are not eligible to vote")
    return wrapper

@check_voter
def voter(age):
    print(f"You are eligible to vote at age {age}")

voter(21)
```

```python
# we have to check given string is email or not
def check_string_is_email(func):
    def wrapper(input_str):
        if "@" in input_str and "." in input_str and " " not in input_str:
            func(input_str)
        else:
            print("Given string is not email")
    return wrapper

@check_string_is_email
def validator(input_str):
    print("Given string is email")

validator("prakash@abc.com")
```
