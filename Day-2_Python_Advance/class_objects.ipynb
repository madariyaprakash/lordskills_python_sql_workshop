***Constructor (__init__)***
- Special method called automatically when you create an object.
- Used to initialize attributes of the class.
- Defined with `def __init__(self, ...)`


```python
class Student:
    #class variable
    school_name = "ABC school"

    #constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # instance method
    def display(self):
        self.age += 10
        print(f"Name: {self.name}, Age: {self.age}")

    @classmethod
    def change_school_name(cls, new_school):
        cls.school_name = new_school
        print(cls.school_name)

s1 = Student("prakash",35)
s1.display()

s1.change_school_name("XYZ school")

```

***Class Method***
- Operates on the class itself, not individual objects.
- Declared with `@classmethod`.
- First parameter is `cls` (the class).
- Often used for alternative constructors or class-level operations.
- Class methods can access the static method but can't access the instance method as it doesn't have `self` reference.

***Static Method***
- Declared with `@staticmethod`.
- Doesn’t take `self` or `cls`.
- Behaves like a normal function but lives inside the class namespace.
- Used when logic is related to the class but doesn’t need object or class state.


```python
# @staticmethod
#     def is_pass(marks):
#         return marks >=45


class Student:
    #class variable
    school_name = "ABC school"

    #constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # instance method
    def display(self):
        self.age += 10
        print(f"Name: {self.name}, Age: {self.age}")

    @classmethod
    def change_school_name(cls, new_school):
        cls.school_name = new_school
        print(cls.school_name)

    @staticmethod
    def check_age(age):
        return age > 18

s1 = Student("prakash",35)
s1.display()

s1.change_school_name("XYZ school")
s1.check_age(35)

```
