# File Handling
-  file handling refers to working with files such as creating, opening, reading, writing, and closing them. 
- It’s essential for storing data permanently, processing external files like .txt, .csv, or .json, and automating tasks such as saving logs or reading configurations.

***Key Functions***
- open(filename, mode) → Opens a file.
- read() / readline() / readlines() → Read data from a file.
- write() / writelines() → Write data to a file.
- close() → Close the file.
- with open(...) as f: → Context manager that automatically closes the file


***File Modes***
- "r" - read(default). Error if file doesn't exist.
- "w" - write. create file if not exists, overwrites if exists.
- "a" - Append. Creates file if not exists, add content at the end.
- "x" - Create. Error if file already exists.
- "t" - Text mode (default)
- "b" - Binary mode (eg. image, pdf)


```python
# writing to a file
with open('output.txt', 'x') as file:
    file.write("Name: James, Marks: 85\n")
    file.write("Name: Bob, Marks: 95\n")
```

```python
# reading from a file
with open("output.txt", "r") as file:
    result = file.readlines()
    print(result)
```

```python
with open("output.txt", "a") as file:
    file.write("Name: bob, Marks: 65\n")
    file.write("Name: john, Marks: 55\n")
```

```python
# appending to a file
with open("output.txt", "a") as file:
    file.writelines(["Name: prakash, Marks: 65\n", "Name: suresh, Marks: 65\n"])
```

```python
# reading line by line
with open("output.txt", "r") as file:
    for line in file:
        print(line.strip())
```

# Working with CSV
Working with CSV (Comma-Separated Values) files in Python is very common for handling tabular data like student records, sales logs, or datasets. Python provides a built-in csv module that makes this easy.

- Import the csv module
- Open the file using open()
- Use csv.reader or csv.writer depending on whether you’re reading or writing
- Always close the file (or better, use with open(...) so it closes automatically
-  `writer = csv.writer(f)`
- `writer.writerow`


```python
import csv

#opening the file
with open("test.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Marks"])
    writer.writerow(["James", 85])
    writer.writerow(["Bob", 95])
```

```python
import csv

# read the data from test.csv
with open("test.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
import csv

# writing student records in csv

with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)

    # header row
    writer.writerow(["Name","Enroll Number","Marks", "Grade"])

    # data row
    writer.writerow(["Prakash", 101, 85, "A"])
    writer.writerow(["Anita", 102, 35, "F"])
    writer.writerow(["Rahul", 103, 92, "A+"])

```

```python
import csv

with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```
