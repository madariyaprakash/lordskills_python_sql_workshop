# Employee Salary Calculation
***instance variable***
 - Create constructor which takes the instance variable name, basic_salary

***methods***
- caluclate_hra() = 20% of basic
- calculate_dra() = 10% of basic
- calculate_total_salary()

```python
class Employee:
    def __init__(self, name, basic_sal):
        self.name =name
        self.basic_sal = basic_sal

    def calculate_hra(self):
        return 0.20*self.basic_sal
    
    def calculate_dra(self):
        return 0.10 * self.basic_sal
    
    def calc_total_sal(self):
        hra = self.calculate_hra()
        dra = self.calculate_dra()
        total_sal = hra+ dra + self.basic_sal
        print("total salary:", total_sal)

emp1 = Employee("prakash", 20000)
emp1.calc_total_sal()
```
