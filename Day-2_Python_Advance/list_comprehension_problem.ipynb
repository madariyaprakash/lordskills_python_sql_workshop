# List Comprehension Problem

1) ***Squares of Even Numbers***

- Write a list comprehension to generate the squares of all even numbers between 1 and 20.
- Expected Output: `[4, 16, 36, 64, 100, 144, 196, 256, 324, 400]`



```python
result = [i**2 for i in range(1,20) if i%2==0]
print(result)
```

`result_list = [operation_on_data  for i in iterable_obj  condition]`

2) ***Filter Words by Length***
- Given a list of words: `["apple", "banana", "kiwi", "grape", "watermelon"]`,
use list comprehension to extract words with more than 5 letters.
Expected Output: `["banana", "watermelon"]`




```python
fruits =["apple", "banana", "kiwi", "grape", "watermelon"]
filtered_fruits = [item for item in fruits if len(item)>5]
filtered_fruits
```

3) ***Multiples of 3***
- Generate a list of numbers between 1 and 50 that are divisible by 3.
- Expected Output: `[3, 6, 9, ..., 48]`




```python
result_divisible_by3 = [i for i in range(1,50) if i%3==0]
result_divisible_by3
```

# Dictionary Comprehension Problem

1) ***Word Lengths***
- Given a list of words: `["python", "java", "c", "ruby"]`,
- create a dictionary mapping each word to its length.
- Expected Output: `{"python":6, "java":4, "c":1, "ruby":4}`

- `syntax - result = {k:v for k,v in dict.items() condition}`


```python
langs = ["python", "java", "c", "ruby"]
result_dict = {item:len(item) for item in langs}
result_dict
```

2) ***Invert Dictionary***
- Given a dictionary `{1:"a", 2:"b", 3:"c"}`,
- use dictionary comprehension to invert it (keys become values, values become keys).
- Expected Output: `{"a":1, "b":2, "c":3}`




```python
d1 = {1:"a", 2:"b", 3:"c"}
invert_dict = {v:k for k, v in d1.items()}
invert_dict
```
